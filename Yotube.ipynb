```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(yotube['comment_count'],yotube['ratings_disabled'], test_size=0.33, random_state=42)
```

```python
lrn = LogisticRegression()
X_train= X_train.values.reshape(-1, 1)
X_test = X_test.values.reshape(-1, 1)
#Reshape data into the correct number of dimensions to be fit.
lrn.fit(X_train, y_train)
```

```python
pred = lrn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, pred)
#3248 True negative, 225 false negative from logistic model.
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
```

```python
pred1 = dtree.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
```

```python

print(confusion_matrix(y_test, pred1))
print('\n')
print(classification_report(y_test, pred1))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(yotube['comment_count'].head(n = 10000),yotube['ratings_disabled'].head(n = 10000), test_size=0.33, random_state=42)
```

```python
X_train= X_train.values.reshape(-1, 1)
X_test = X_test.values.reshape(-1, 1)
rfc = RandomForestClassifier(n_estimators = 200)
rfc.fit(X_train,y_train)
```

```python
rfcpred = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test, rfcpred))
print('\n')
print(classification_report(y_test,rfcpred))
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
cm = confusion_matrix(y_test, rfcpred)
ax.grid(False)
ax.xaxis.set(ticks=(0, 1), ticklabels=('Predicted False', 'Predicted True'))
ax.yaxis.set(ticks=(0, 1), ticklabels=('Actual False', 'Actual True'))
ax.imshow(cm)
plt.plot()
for i in range(2):
    for j in range(2):
        ax.text(j, i, cm[i, j], ha='center', va='center', color='pink', fontsize = 30)
        print(i,j)
cm[0,1]
```
